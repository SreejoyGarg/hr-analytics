```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('../data/hr_attrition.csv')
print("Shape:", df.shape)
df.head()
```

```python
print("Column Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)
```

```python
print("Missing Values:")
print(df.isnull().sum().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nUnique Employee Numbers:", df['EmployeeNumber'].nunique())
print("Total Rows:", len(df))
```

```python
# Convert Attrition to binary (easier for analysis)
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Salary Band (same as our SQL query)
df['Salary_Band'] = pd.cut(
    df['MonthlyIncome'],
    bins=[0, 3000, 6000, 10000, 25000],
    labels=['Low (Below 3K)', 'Mid-Low (3K-6K)', 'Mid-High (6K-10K)', 'High (Above 10K)']
)

# Age Group
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 25, 35, 45, 55, 100],
    labels=['Under 25', '25-34', '35-44', '45-54', '55+']
)

# Combined Risk Flag (Overtime + Low Salary)
df['Risk_Category'] = np.where(
    (df['OverTime'] == 'Yes') & (df['MonthlyIncome'] < 5000),
    'High Risk',
    'Lower Risk'
)

print("New columns added successfully!")
df[['MonthlyIncome', 'Salary_Band', 'Age', 'Age_Group', 'OverTime', 'Risk_Category']].head()
```

```python
df.to_csv('../data/hr_clean.csv', index=False)
print("Cleaned file saved successfully!")
```
